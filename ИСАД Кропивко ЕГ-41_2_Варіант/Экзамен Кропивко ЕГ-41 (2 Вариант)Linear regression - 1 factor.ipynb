### Outliers

```python
df.describe()
```

```python
# Function Outliers
def outliers(df):
    num_var = list(df._get_numeric_data().columns)
    for col_names in num_var:
        df[col_names] = df[col_names].apply(lambda y: df[col_names].mean()-3*df[col_names].std() 
                            if y < df[col_names].mean()-3*df[col_names].std() else y)
        df[col_names] = df[col_names].apply(lambda y: df[col_names].mean()+3*df[col_names].std() 
                            if y > df[col_names].mean()+3*df[col_names].std() else y)
    return(df)
```

```python
# Outliers
df = outliers(df)
df.describe()
```

### Splitting

```python
# Splitting the dataset into the Training set and Test set

f = ['R&D Spend','Administration','Marketing Spend','State']
X = df[f]
y = df['Profit']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

### Linear regression - 1 factor

Для 'у' выбираю 'Profit', делаем предположение что заказчика будут интерессовать именно эти значения, а для 'х' выбираем 'R&D Spend'. Так как у него наибольшая корреляция с 'Profit' и их парное распределение больше всего напоминало прямую.


```python
# Fitting Simple Linear Regression to the Training set (M2)
f1 = ['R&D Spend']

from sklearn.linear_model import LinearRegression
sr = LinearRegression().fit(X_train[f1], y_train)

# Getting parameters
sr.coef_, sr.intercept_
```

```python
# Predicting the Test set results
y_pred = sr.predict(X_test[f1])

# Coefficient of determination R^2
sr.score(X_train[f1], y_train).round(3)
```

```python
# Mean squared error
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train, sr.predict(X_train[f1]), squared=False).round(2), mean_squared_error(y_test, y_pred, squared=False).round(2)
```

```python
# Visualising the Test set results
plt.scatter(X_test['R&D Spend'], y_test, color = 'red')
plt.plot(X_test['R&D Spend'], sr.predict(X_test[f1]), 'bo')
plt.title('Profit vs R&D Spend (Test set)')
plt.xlabel('R&D Spend')
plt.ylabel('Profit')
plt.show()
```

Делам выводи того, что модель достаточно качественная, так как имеет большой R^2. Средняя квадратическая ошибка тестовой и тренировочной выборки весомо разнятся. Модель не переобучена.
